```python
import tensorflow as tf
import numpy as np
import os
import struct
import matplotlib.pyplot as plt
from PIL import Image

# ============================================================
# 1. NASTAVITVE
# ============================================================
BITMAP_DIR = r'F:\FRI_DIPLOMSKA\-delovna mapa-\DATASET\sken-OBRAZEC\vse'
TEST_OSEBE = ['Dokument_2026-05-24_113242', '2026-05-24_113215']
FILTER_CRKE = None

# ============================================================
# 2. NALOŽI PODATKE
# ============================================================
X_all, y_all, osebe_all = [], [], []

for filename in os.listdir(BITMAP_DIR):
    if not filename.lower().endswith('.png'):
        continue

    # mal drgačno poimenovanje slik kle
    # Vzemi ime brez končnice .png
    ime_brez_png = filename.rsplit('.', 1)[0]
    parts = ime_brez_png.split('_')
    
    crka = parts[0] # Prva komponenta je vedno črka (npr. 'M')
    
    # Združimo vse vmesne dele, ki sestavljajo ime dokumenta/osebe
    celotno_ime = '_'.join(parts[1:])
    
    # OČISTIMO IME OSEBE: Odstranimo "_celicaXX" ali oklepaje na koncu, če obstajajo
    # To zagotovi, da "Dokument_... (8)" postane čisto "Dokument_..."
    if ' (' in celotno_ime:
        oseba = celotno_ime.split(' (')[0]
    elif '_celica' in celotno_ime:
        oseba = celotno_ime.split('_celica')[0]
    else:
        # Če je na koncu le zadnji del (npr. celica01) brez posebne oznake
        oseba = '_'.join(parts[1:-1]) if len(parts) > 2 else 'unknown'



    if FILTER_CRKE and crka not in FILTER_CRKE:
        continue

    img = Image.open(os.path.join(BITMAP_DIR, filename)).convert('L')
    # S tem spremeniš belo ozadje v črno in črno pisalo v belo
    pixels = 1.0 - (np.array(img, dtype=np.float32) / 255.0)
    #pixels = np.array(img, dtype=np.float32) / 255.0  # [0, 1]

    X_all.append(pixels.flatten())  # 28x28 -> 784
    y_all.append(crka)
    osebe_all.append(oseba)

X_all = np.array(X_all)
y_all = np.array(y_all)
osebe_all = np.array(osebe_all)

print(f"Skupaj vzorcev: {len(X_all)}")
print(f"Črke: {sorted(set(y_all))}")
print(f"Osebe: {sorted(set(osebe_all))}")

# ============================================================
# 3. TRAIN/TEST SPLIT PO OSEBAH
# ============================================================
train_mask = ~np.isin(osebe_all, TEST_OSEBE)
test_mask  =  np.isin(osebe_all, TEST_OSEBE)

X_train = X_all[train_mask]
y_train = y_all[train_mask]
X_test  = X_all[test_mask]
y_test  = y_all[test_mask]

print(f"\nTrain: {X_train.shape} — {np.unique(osebe_all[train_mask])}")
print(f"Test:  {X_test.shape} — {np.unique(osebe_all[test_mask])}")





labels_ordered = ['A', 'B', 'C', 'Č', 'D', 'E', 'F', 'G', 'H', 'I',
                  'J', 'K', 'L', 'M', 'N', 'O', 'P', 'R', 'S', 'Š',
                  'T', 'U', 'V', 'Z', 'Ž']
label_to_idx = {l: i for i, l in enumerate(labels_ordered)}
y_train_enc = np.array([label_to_idx[l] for l in y_train])
y_test_enc  = np.array([label_to_idx[l] for l in y_test])


# ============================================================
# 4. VIZUALIZACIJA VZORCEV
# ============================================================
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i in range(min(10, len(X_all))):
    axes[i//5][i%5].imshow(X_all[i].reshape(28, 28), cmap='gray')
    axes[i//5][i%5].set_title(f'{y_all[i]} ({osebe_all[i]})')
    axes[i//5][i%5].axis('off')
plt.suptitle('Vzorci obrazca')
plt.tight_layout()
plt.show()

# ============================================================
# 5. MODEL
# ============================================================
n_classes = len(labels_ordered)

bitmap_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(784,)),
    tf.keras.layers.Reshape((28, 28, 1)),

    tf.keras.layers.SeparableConv2D(16, (3,3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2,2)),

    tf.keras.layers.SeparableConv2D(32, (3,3), padding='same', activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2,2)),

    #tf.keras.layers.Flatten(),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(n_classes, activation='softmax')
])

bitmap_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])
bitmap_model.summary()

# ============================================================
# 6. TRENING
# ============================================================
early_stop = tf.keras.callbacks.EarlyStopping(
    patience=10,
    restore_best_weights=True,
    verbose=1
)

history = bitmap_model.fit(
    X_train, y_train_enc,
    epochs=100,
    batch_size=16,
    validation_data=(X_test, y_test_enc),
    callbacks=[early_stop],
    verbose=1
)

# ============================================================
# 7. REZULTATI
# ============================================================
_, train_acc = bitmap_model.evaluate(X_train, y_train_enc, verbose=0)
_, test_acc  = bitmap_model.evaluate(X_test,  y_test_enc,  verbose=0)
print(f"\nTrain accuracy: {train_acc*100:.1f}%")
print(f"Test accuracy:  {test_acc*100:.1f}%")

# Graf
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Test')
plt.title('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Test')
plt.title('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# 8. PRETVORBA V TFLITE (z optimizacijo za mikrokontrolerje)
# ============================================================
converter = tf.lite.TFLiteConverter.from_keras_model(bitmap_model)

# Osnovna pretvorba
tflite_model = converter.convert()
with open(r'F:\FRI_DIPLOMSKA\-delovna mapa-\MODEL\bitmaps\obrazec.tflite', 'wb') as f:
    f.write(tflite_model)

# ŠE BOLJE: Kvantizacija (INT8 ali Float16) za hitrejši STM32
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_quant_model = converter.convert()
with open(r'F:\FRI_DIPLOMSKA\-delovna mapa-\MODEL\bitmaps\obrazec2.tflite', 'wb') as f:
    f.write(tflite_quant_model)

print("TFLite modela (navaden in kvantiziran) shranjena!")
```
